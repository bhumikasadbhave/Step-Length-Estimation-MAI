```python
def take_mean(acc):
    arr=np.array(acc)
    mean_acc= arr.mean()    
    return mean_acc

def take_median(acc):
    arr=np.array(acc)
    median_acc= np.median(arr)    
    return median_acc

def histogram_features(data):
    hist, _ = np.histogram(data, bins=50)
    single_feature = np.mean(hist)
    return single_feature


def data_around_medians(data):
    median_value = np.median(data)
    sorted_list = sorted(data)
    median_value_org = min(sorted_list, key=lambda x: abs(x - median_value))
    median_index = sorted_list.index(median_value_org)
    closest_numbers = sorted_list[median_index - 5:median_index + 5]
    mean_acc= np.mean(closest_numbers)  
    return mean_acc

def middle_elements(data):
    middle_index = len(data) // 2
    middle_elements = data[middle_index - 10 : middle_index + 10]
    mean_acc= np.mean(middle_elements)  
    return mean_acc

def histogram_weights(data):
    weights = np.ones_like(data)
    middle_index = len(data) // 2
    middle_elements = (data[middle_index - 5 : middle_index + 5])
    middle_range = np.array([True if x in middle_elements else False for x in data])
    weights[middle_range]= 2
    weights[middle_range==False]=0.5
    hist, _ = np.histogram(data, bins=20, weights=weights)
    single_feature = np.mean(hist)
    return single_feature

def histogram_all_clm(data):
    hist, _ = np.histogram(data, bins=100,density=True)
    return hist
    

for col in columns:
    data[col]=data[col].apply(histogram_weights)
    test_1[col]=test_1[col].apply(histogram_weights)
    test_2[col]=test_2[col].apply(histogram_weights)
    test_3[col]=test_3[col].apply(histogram_weights)


remove_columns = ['accelerationMagnitudeLeft','accelerationMagnitudeRight','stepStartIndex','stepStartEndIndex', 'stepEndStartIndex', 'stepEndEndIndex','Height','Shoes']
data = data.drop(remove_columns,axis=1)
test_1= test_1.drop(remove_columns,axis=1)
test_2= test_2.drop(remove_columns,axis=1)
test_3= test_3.drop(remove_columns,axis=1)
```

```python
def SD(y_truth,y_pred):
    residuals = y_pred - y_truth
    squared_residuals = residuals**2
    mean_squared_residuals = np.mean(squared_residuals)
    std_dev = np.sqrt(mean_squared_residuals)
    return std_dev
```
